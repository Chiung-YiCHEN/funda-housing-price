```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import re
import time

idnumber = []
for i in range(1, 1, 1):

    page = 'https://www.funda.nl/koop/amsterdam/0-400000/+15km/p'+ \
        str(i)
    web_result = requests.get(page).text
    soup = BeautifulSoup(web_result, "lxml")
    idnumber.extend([item['data-save-object-tinyid']
                        for item in soup.find_all(attrs={"data-save-object-tinyid": True})])
    

#for link in soup.findAll('a', attrs={'href': re.compile("http://")}):
#    links.append(link.get('href'))
#print(links)
```
